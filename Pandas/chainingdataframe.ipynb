```python
# just a one function for almost all of the preoricessing of features. Everything we did above 
# can be summarized in this function, it helps the notebook clean and well-expllanatory everthing

def preprocess_df(df):
    cols = ['city08', 'comb08', 'highway08', 'cylinders','displ','drive','eng_dscr',
        'fuelCost08','make','model','trany','range','createdOn','year']
    return(
    df
    [cols]
    .assign(cylinders = df['cylinders'].fillna(0).astype('int8'),
            displ = df['displ'].fillna(0).astype('float16'),
            drive = df['drive'].fillna('Other').astype('category'),
            automatic = df['trany'].str.contains('Auto'),
            speed = df['trany'].str.extract(r'(\d)+').fillna('20').astype('int8'),
            createdOn = pd.to_datetime(df.createdOn.replace({'EDT':'-04:00',
                        'EST':'-05:00'}, regex = True)),
            ffs = df['eng_dscr'].str.contains('FFS')
            )
    .astype(
        {
            'highway08':'int8','city08':'int16','comb08':'int16','fuelCost08':'int16',
            'range':'int16','year':'int16','make':'category'
        }
    )
    .drop(columns=['trany','eng_dscr'])
    
)
```

```python
df1 = preprocess_df(df)
```

**For debugging in chaining you can just comment out the undesired code or just use pipe function**


```python
def getIntermediateVariable(df, var_name):
    globals()[var_name] = df
    return df
    
def tweak_df(df):
    cols = ['city08', 'comb08', 'highway08', 'cylinders','displ','drive','eng_dscr',
        'fuelCost08','make','model','trany','range','createdOn','year']
    return(
    df
    [cols]
    
    # get this value
    .pipe(getIntermediateVariable, 'df3')
    .assign(cylinders = df['cylinders'].fillna(0).astype('int8'),
            displ = df['displ'].fillna(0).astype('float16'),
            drive = df['drive'].fillna('Other').astype('category'),
            automatic = df['trany'].str.contains('Auto'),
            speed = df['trany'].str.extract(r'(\d)+').fillna('20').astype('int8'),
            createdOn = pd.to_datetime(df.createdOn.replace({'EDT':'-04:00',
                        'EST':'-05:00'}, regex = True)),
            ffs = df['eng_dscr'].str.contains('FFS')
            )
    .astype(
        {
            'highway08':'int8','city08':'int16','comb08':'int16','fuelCost08':'int16',
            'range':'int16','year':'int16','make':'category'
        }
    )
    .drop(columns=['trany','eng_dscr'])
    
)
    
```
